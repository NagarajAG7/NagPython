**Common Errors**

- Unicode error: due to single slash
    - provide double slash or apply r (raw string)

- file not found error
    - either wrong location or spell mistake

- permission denied error
    - due to the file is already opened
    - close that file and read it again


```python
open('C:\\Users\\Nagaraj AG\\OneDrive\\Desktop\\Temp Files\\Python Assignment - 4Jan2025.txt').read()
```

```python
# method 2
with open(filename) as file:
    data2=file.read()
print(data2)
```
